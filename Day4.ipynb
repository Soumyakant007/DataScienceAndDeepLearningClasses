# Visualising Time Series Data

    -Plot simple time series plots
    -Derive variables such as month and year and use them for richer visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("white")
```

```python
market_df=pd.read_csv("./global_sales_data/market_fact.csv")
customer_df=pd.read_csv("./global_sales_data/cust_dimen.csv")
product_df=pd.read_csv("./global_sales_data/prod_dimen.csv")
shipping_df=pd.read_csv("./global_sales_data/shipping_dimen.csv")
orders_df=pd.read_csv("./global_sales_data/orders_dimen.csv")
```

```python
market_df.head()
```

```python
orders_df.head()
```

```python
df=pd.merge(market_df,orders_df,how="inner",on="Ord_id")
df.shape
```

```python
product_df.head()
```

```python
df.info()
```

```python
print(df["Order_Date"])
```

```python
df["Order_Date"]=pd.to_datetime(df["Order_Date"])
```

```python
time_df=df.groupby("Order_Date")["Sales"].sum()
print(time_df.head())
```

```python
type(time_df)
```

```python
plt.figure(figsize=(16,3))

sns.lineplot(data=time_df)
```

## Using Derived Date Metrics of Visualization

##### It is often helped to use derived variables from date such as month and year and using them to identify hidden patterns

```python
df["month"]=df["Order_Date"].dt.month
df["year"]=df["Order_Date"].dt.year
df.head()
```

```python
df_time=df.groupby(["year","month"]).Sales.mean()
df_time.head()
df_time.shape
```

```python
df_time = pd.DataFrame({
    'time': pd.date_range(start='1/1/2020', periods=100, freq='D'),
    'sales': np.random.rand(100)
})
plt.figure(figsize=(12,8))
sns.lineplot(data=df_time)
```

```python
year_month=pd.pivot_table(df,values="Sales",index="year",columns="month",aggfunc="mean")
year_month.head()
```

```python
help(sns.heatmap)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(year_month,cmap="YlGnBu")
```
