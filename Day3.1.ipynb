```python
plt.plot(x,y,"bo")
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm law")

plt.xlim([20,80])
plt.ylim([200,800])
plt.show()
```

```python
x=np.linspace(0,10,20)
y=x*2
plt.plot(x,y,"+")
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm law")
plt.show()
```

```python
x=np.linspace(0,5,10)
y=np.linspace(3,6,10)

plt.plot(x,y,"r>",x,y**2,"b+",x,y**3,"c*")
plt.show()
```

# Figures and SubPlot

```python
x=np.linspace(1,10,100)
y=np.log(x)

plt.figure(0)
#create subplot with 1 row and 2 column
plt.subplot(3,1,1) #equivalent to plt.subplot(1,2,1)
plt.title("y=log(x)")
plt.plot(x,y)
plt.show()

plt.subplot(3,1,2)
plt.title("y=y**2")
plt.plot(x,y**2)
plt.show()

plt.subplot(3,1,3)
plt.title("y=y*2")
plt.plot(x,y*2)
plt.show()
```

```python
import math
x=np.linspace(1,10,100)

plt.figure(1)

plt.subplot(2,2,1)
plt.title("linear")
plt.plot(x,x)

plt.subplot(2,2,2)
plt.title("cubic")
plt.plot(x,x**3)

plt.figure(2)
plt.subplot(2,2,1)
plt.title("log")
plt.plot(x,np.log(x))

plt.subplot(2,2,2)
plt.title("exponential")
plt.plot(x,math.e**x)
```
